Tumblr è stato lanciato nel febbraio 2007 e nel giro di due settimane il servizio ha raggiunto 75.000 utenti.

Nel dicembre 2018, alcune settimane dopo l'eliminazione dell'app per Tumblr dall'App Store, la piattaforma ha comunicato di essere intenzionata a eliminare tutti i contenuti ritenuti per adulti, per via di critiche sugli standard applicati nel controllo dei contenuti. Queste restrizioni hanno contemporaneamente suscitato le proteste di molti utenti, poiché a loro dire andrebbero a penalizzare ingiustamente contenuti artistici o scientifici. È stato calcolato che, in seguito a questa politica, nei primi mesi del 2019 Tumblr abbia perso 437 milioni di visite totali alle proprie pagine.
Ad agosto 2019, Verizon cede Tumblr ad Automattic, società che controlla il Cms web open source WordPress.</p>

In questi anni Tumblr è stato utile per molti ragazzi che hanno avuto la possibilità condividere pensieri intimi con il resto del mondo senza esserne giudicati. Questo è possibile perchè non è presente il concetto di amicizia all'interno della piattaforma, dunque, i lettori del proprio blog saranno persone di tutto il mondo che condividono la stessa passione o interesse.</p>

    <h2>API Rest: cosa sono?</h2>
    <br>
</div>
<div style = "text-align: left; ">
    <h2>Cosa si intende con la parola Rest?</h2>
    <p>
Quando si parla di Rest (Representational State Transfer) ci si riferisce a un’architettura software. Questo approccio architetturale venne ideato per creare web API basandosi sul protocollo HTTP.

Il REST è infatti un sistema di trasmissione dei dati che utilizza l’HTTP e che fa un grande uso delle funzioni di HTTP, ricorrendo ai comandi GET, POST, PUT, PATCH, DELETE e OPTIONS grazie ai quali si riesce ad identificare ben precise risorse presenti in tutto il web.</p>
    <h2>I princìpi del REST</h2>
    <p>

Il REST si compone da una serie di vincoli e princìpi:

• Client-server: essi hanno compiti tra loro ben definiti e dedicati. Ognuno di loro si occupa di ben determinati aspetti e funzionalità, evitando così qualsiasi accavallamento di competenze;

• Stateless: questo principio si basa sull’assenza di stato durante la comunicazione tra server e client. Questo comporta che ogni sessione client-server è unica, mai correlata alle precedenti o a quelle future;

• Cacheable: questo principio permette di memorizzare i dati forniti dal server all'interno di una cache. Le risposte date dal server devono essere esplicitamente etichettate come memorizzabili nella cache o meno: in questo modo si evita che i client usino stati vecchi o addirittura errati;

• Layered system: Rest è un sistema basato su livelli composti da server distribuiti che garantiscono i vari servizi sopracitati.

• Uniform interface: interfaccia dei server e dei client è uniforme e omogenea, facile da utilizzare e perfetta se si vuole semplificare l’architettura;

• Code on demand: Rest consente di aggiornare la logica dei client senza dover aggiornare obbligatoriamente anche quella dei server.
</p>
    <h2>API Rest: perché usarle?</h2>
    <p>

L’API REST, sono delle API che seguono i princìpi Rest e che per funzionare ricorrono ai comandi dell’HTTP sopracitati. I dati di ritorno saranno in JSON.
    </p>
    <h4>I vantaggi dell’utilizzo delle API Rest</h4>
<p>
• Indipendenza: le API Rest sono indipendenti dai linguaggi o da precise piattaforme. Si possono avere server che utilizzano qualsiasi linguaggiio di programmazione senza riscontrare alcun problema nel loro utilizzo.

• Separazione client-server: grazie a tale separazione si possono trattare indipendentemente le diverse componenti. Dunque, è modificare solo una parte progettuale senza, per questo, essere obbligati a mettere mano sia al server che al client. 

• Scalabilità: la separazione client-server si traduce in una miglior scalabilità del sistema stesso.
</p>
    <p>In conclusione possiamo utilizzare le API Rest per richiedere delle risorse ad un server definendo un URI (sequenza di caratteri che identifica universalmente ed univocamente una risorsa).
    </p>
  
</div>

Il segmento successivo del percorso dell'URI dipende dal tipo di richiesta che si desidera effettuare.
    Vediamo insieme alcuni metodi che possono essere inseriti nelle richieste:</p>
    <ul>
        <li>/info: questo metodo restituisce informazioni generali sul blog, come il titolo, la descrizione, se è possibile seguirlo o contattarlo, l'immagine di profilo, il numero di post e altri dati di alto livello.</li>

        <li>/following: questo metodo ritorna i blog seguiti.</li>
        <li>/followers: questo metodo ritorna i follower del blog.</li>

    </ul>

    <h2>API Wrapper</h2>
    <p>Oltre alle chiamate API Rest, Tumblr mette a disposizione una serie di wrapper che forniscono una serie di metodi utili per l'utilizzo delle API. Tra i vari linguaggi per cui è stato creato un wrapper abbiamo Python, PHP, JAVA, Javascript, Ruby, GO e Objective-C</p>
    <h2>Cos'è un API wrapper?</h2>
    <p>Un wrapper API è un pacchetto specifico per un determinato linguaggio di programmazione, esso incapsula le chiamate API in dei metodi utilizzabili all'interno del codice rendendo semplici funzioni complicate. Dunque, esso aiuta gli sviluppatori a chiamare varie API senza la necessità di andare a definire le varie chiamate.</p>
    <h2>Pytumblr</h2>
    <p>Tra i vari wrapper messi a disposizione da Tumblr c'è Pytumblr, che come si può intuire dal nome è il wrapper per Python,  esso permette di effettuare chiamate API andando ad utilizzare dei semplici metodi che ritornano le informazioni richieste.</p>

    <p>Essi si suddividono in 2 categorie:</p>
        <ul>
            <li><b>User methods:</b> utili per richiedere informazioni riguardanti le info dell'utente che sta effettuando le chiamate API:
                <ul>
                    <li>.info(): ritorna le informazioni riguardanti l'utente</li>
                    <li>.dashboard(): ritorna la dashboard</li>
                    <li>.likes(): ritorna i <i>mi piace</i> messi</li>
                    <li>.following(): ritorna i blog <i>seguiti</i> dall'utente</li>
                    <li>.follow('#blog'): permette di <i>seguire</i> il blog passato al metodo</li>
                    <li>.unfollow('#blog'): permette di smettere di <i>seguire</i> il blog passato al metodo</li>
                    <li>.like(#id_post, 'reblog_key')permette di mettere <i>mi piace</i> al post identificato dall'id passato</li>
                    <li>.unlike(#id_post, 'reblog_key'): permette di rimuovere il <i>mi piace</i> al post identificato dall'id passato</li>
                </ul>
            </li>
            <br>
            <li><b>Blog methods:</b> utili per richiedere informazioni riguardanti le info di un determinato blog:
                <ul>
                    <li>.blog_info('#blog_name'): ritorna le informazioni riguardanti il blog inserito</li>
                    <li>.posts('#blog_name',#param): ritorna i post publicati dal blog specificato, si possono inoltre aggiungere dei parametri come ad esempio <i>reblog_info=True</i> e <i>notes_info=True</i> che permettono di ricevere delle informazioni riguardanti le note relative ai singoli post e le informazioni riguardanti i reblog effettuarti sul post</li>
                    <li>.avatar('#blog_name'): ritorna l'avatar del blog inserito</li>
                    <li>.blog_likes('#blog_name'): ritorna i mi piace del blog inserito</li>
                    <li>.followers('#blog_name'): ritorna i followers del blog inserito </li>
                    <li>.blog_following('#blog_name'): ritorna i i follow pubblici del blog inserito</li>
                    <li>.create_text('#blog_name', state="published", slug="testing-text-posts", title="#title", body="#text"): permette di creare un nuovo post di tipo testo, esso avrà titolo e body uguale a quello inserito nella chiamat API </li>
                    <li>.delete_post('#blog_name',#id_post): esegue la cancellazione del post con id uguale a quello specificato nella chiamata.</li>
                </ul>            
            </li>
        </ul>

```python
###################################################Estrazione dei dati##########################################################
import pytumblr
import pandas as pd
import json

# Authenticate via OAuth
client = pytumblr.TumblrRestClient(
  'GTWFIFiJKuQYG8S9Xl5HfFCUQi5oqr7lEQwupIcgEaG6fCJCPr',
  '37FOyIrFl3rxSh6dmg6XhGx1TH2IzSc1hjSkJnxSMgFC9W0kQ3',
  '1HzYwuuvY2Ggbl2lVBtuN145SGTtaPJXKTFtrEtoLGA7G4Jvtd',
  '0SrIA1TfRejIhVQxeJEla7F9UMJp5TMeXoybuDHO8T7vWsqRQS'
)

#blog da cui si stanno estraendo le informazioni
blog="sergioct98"

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##lettori del blog
#print("Lettori del blog\n")
followers=client.followers(blog)
followers_number=followers["total_users"]
#print(json.dumps(followers, sort_keys=True, indent=4))

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##blog seguiti
#print("Blog seguiti\n")
follow=client.following()
follow_number=follow["total_blogs"]
#print(json.dumps(follow, sort_keys=True, indent=4))

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##post pubblicati
#print("Post pubblicati\n")
posts_number=client.posts(blog, reblog_info=True, notes_info=True)["total_posts"]

#controllo limite massimo
if posts_number >=20:
    posts_number=19
    
post_list=[]
for index in range(posts_number):
        post_list.append(client.posts(blog, reblog_info=True, notes_info=True)['posts'][index])
#print(json.dumps(post_list, sort_keys=True, indent=4))

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##Estraiamo le tipologie dei post
#print("Tipologie dei post\n")
type_post_list=[]
#tabella con tipologia di post e frequenze, essa verrà utilizzata per mostrare graficamente i dati
type_post_table = pd.DataFrame(columns=['Type','Count'])
for post in post_list:
    type_post=post["type"]
    type_post_list.append(type_post)

#set() utilizzato per rendere univoci i valori della lista passata
type_post_unique=set(type_post_list)

for index in range(len(type_post_unique)):
    type_tmp=type_post_unique.pop()
    type_post_table=type_post_table.append({'Type':type_tmp,'Count':type_post_list.count(type_tmp)},ignore_index=True)
    
#print(type_post_table)

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##Estraiamo le interazioni ricevute nei post dividendole per tipologia
#print("Interazioni ricevute\n")
likes_list=[]
reblog_list=[]
reply_list=[]
notes_total=0
for post in post_list:
    notes_number=post["note_count"]

    #controllo limite massimo
    if notes_number>=20:
        notes_number=19
    
    notes_total=notes_total+notes_number
    for noteIndex in range(notes_number):
        typeAction=post["notes"][noteIndex]["type"]
        if typeAction=="like":
            likes_list.append(typeAction)
        elif typeAction=="reblog":
            reblog_list.append(typeAction)
        else :
            reply_list.append(typeAction)

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##Estraiamo i tag utilizzati nei post contandone la frequenza
import collections

#print("Tag utilizzati dall'utente\n")

#tabella con tipologia di post e frequenze, essa verrà utilizzata per mostrare graficamente i dati
tags_table = pd.DataFrame(columns=['Tag','Count'])
#lista con tutti i tag inseriti dall'utente
tags_post_list=[]
for post in post_list:
    tags_number=len(post["tags"])
    
    #controllo limite massimo
    if tags_number>=20:
        tags_number=19
        
    for tagIndex in range(tags_number):
        tags_post_list.append(post["tags"][tagIndex])

#set() utilizzato per rendere univoci i valori della lista passata
tags_unique=set(tags_post_list)

for index in range(len(tags_unique)):
    tag_tmp=tags_unique.pop()
    tags_table=tags_table.append({'Tag':tag_tmp,'Count':tags_post_list.count(tag_tmp)},ignore_index=True)

tags_table=tags_table.sort_values(by=['Count'],ascending=False)
#print(tags_table)

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")

##Estraiamo le interazioni che l'utente ha effettuato con gli altri blog
#print("Post a cui l'utente ha messo mi piace \n")
blog_likes=client.likes()
#print(json.dumps(blog_likes, sort_keys=True, indent=4))

#print("\n-----------------------------------------------------------------------------------------------------------------------\n")
```

```python
###############################Manipolazione dei dati estratti e conclusioni finali###############################################

#Poiché l'obiettivo dell'esempio è quello di estrarre il blog con cui l'utente ha avuto più interazioni, 
#bisogna raffinare le informazioni che sono già in nostro possesso.

#Lista dei blog che l'utente segue
print("\nBlog che l'utente segue: \n")
follow_name=[]
for blogs in follow["blogs"]:
    follow_name.append(blogs["name"])
print(follow_name)

print("\n-----------------------------------------------------------------------------------------------------------------------\n")

#Interazioni che l'utente ha avuto con i blog che segue
print("Blog seguiti a cui l'utente ha messo mi piace ad almeno un post:\n")
blog_notes=[]
for post in blog_likes["liked_posts"]:
    if post["followed"]==True:
        blog_notes.append(post["blog"]["name"])
print(blog_notes)

print("\n-----------------------------------------------------------------------------------------------------------------------\n")

print("Tabella rappresentante il numero di interazioni per ogni blog seguito:\n")

#tabella con blog e mi piace messi
likes_table = pd.DataFrame(columns=['Blog','#Likes'])
header=["Blog","#Likes"]

#rendiamo univoci i blog con cui l'utente ha avuto interazioni
likes_unique=set(blog_notes)

for index in range(len(likes_unique)):
    blog_tmp=likes_unique.pop()
    likes_table=likes_table.append({'Blog':blog_tmp,'#Likes':blog_notes.count(blog_tmp)},ignore_index=True)

likes_table=likes_table.sort_values(by=['#Likes'],ascending=False)

print(tabulate(likes_table, headers=header, tablefmt="grid", showindex=False))

print("\n-----------------------------------------------------------------------------------------------------------------------\n")
#Una volta creata e ordinata la tabella contenente i vari blog seguiti con cui l'utente ha interagito, basterà estrarre il primo
#blog per ottenere il blog più affine.
blog_affine=likes_table.max()
print( blog, "ha maggiore affinità con il blog", blog_affine["Blog"])
```
